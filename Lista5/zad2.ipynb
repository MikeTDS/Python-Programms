```python
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
```

```python
m = 9019
```

Moje oceny:

```python
my_ratings = np.zeros((m,1))
my_ratings[2571] = 5 #Matrix
my_ratings[32] = 4 #Twelve Monkeys
my_ratings[260] = 5 #Star Wars
my_ratings[1097] = 4
my_ratings = np.nan_to_num(my_ratings/np.linalg.norm(my_ratings, axis=0)) #normalizacja macierzy
```

Importowanie danych

```python
X = [np.zeros(m)]
current_user_id=1
with open('./ml-latest-small/ratings.csv', 'r') as file:
    file.readline()
    for line in file:
        line = line.split(',')
        if int(line[1]) < m:
            if int(line[0]) == current_user_id:
                X[current_user_id-1][int(line[1])] = float(line[2])
            else:
                current_user_id += 1
                X.append(np.zeros(m))
                X[current_user_id-1][int(line[1])] = float(line[2])
X
```

Normalizacja macierzy X

```python
X = np.array(X)
X = np.nan_to_num(X/np.linalg.norm(X, axis=0))
```

Obliczanie macierzy Z oraz rekomendacji

```python
Z = np.dot(X,my_ratings)
Z = Z/np.linalg.norm(Z)
Res = np.dot(X.T,Z)
```

Lista posortowanych wyników

```python
t = list(Res)
Recommendations = []
for i in range(len(t)):
    Recommendations.append((t[i][0],i))
Recommendations.sort(reverse=True)
Recommendations
```

Importowanie i wyświetlanie tytułów filmów według rekomendacji

```python
movies = dict()
with open('./ml-latest-small/movies.csv', 'r') as file:
    file.readline()
    for movie in file:
        movie = movie.split(',')
        if int(movie[0]) < m:
            movies.update({int(movie[0]) : movie[1]})
        else:
            break
```

```python
for r in Recommendations:
    if movies.get(int(r[1])) is not None:
        #print(r[1],movies.get(int(r[1]))) #zostaną wypisane id + tytuł
        print(movies.get(r[1])) #sam tytuł
```
