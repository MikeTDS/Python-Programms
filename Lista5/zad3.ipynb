```python
import numpy as np
from scipy.sparse import lil_matrix, csr_matrix
from scipy.sparse.linalg import norm
from sklearn.preprocessing import normalize
np.seterr(divide='ignore', invalid='ignore')
```

```python
m = 193886
n = 283228
```

```python
my_ratings = lil_matrix((m,1))
my_ratings[2571] = 5 #Matrix
my_ratings[32] = 4 #Twelve Monkeys
my_ratings[260] = 5 #Star Wars
my_ratings[1097] = 4
```

```python
X = lil_matrix((n,m), dtype=float)
```

```python
current_user_id=1
with open('./ml-latest/ratings.csv', 'r') as file:
    file.readline()
    for line in file:
        line = line.split(',')
        if int(line[1]) < m:
            if int(line[0]) == current_user_id:
                X[current_user_id-1,int(line[1])] = float(line[2])
            else:
                current_user_id += 1
                X[current_user_id-1,int(line[1])] = float(line[2])
                #print(current_user_id)
```

```python
my_ratings = np.nan_to_num(normalize(my_ratings, norm='l2', axis=0))
X = np.nan_to_num(normalize(X, norm='l2', axis=0))
Z = np.nan_to_num(np.dot(X, my_ratings))
Z = np.nan_to_num(normalize(Z, norm='l2', axis=0))
```

```python
Recommendation = np.dot(X.T, Z).toarray()
Recommendation = np.concatenate((Recommendation, np.array(range(0,m))[:,None]), axis=1)
Recommendation = Recommendation[Recommendation[:,0].argsort()]
Recommendation = Recommendation[::-1]
```

```python
movies = dict()
with open('./ml-latest/movies.csv', 'r') as file:
    file.readline()
    for movie in file:
        movie = movie.split(',')
        if int(movie[0]) < m:
            movies.update({int(movie[0]) : movie[1]})
        else:
            break
```

```python
for r in Recommendation:
    mov = movies.get(int(r[1]))
    if mov:
        #print(r[0],int(r[1]), movies.get(int(r[1]))) #ocena, id, tytuł
        print(movies.get(int(r[1]))) #tytuł
```
